```python
def slo(file, product):

    import re
    import parse
    import pdfplumber
    import pandas as pd
    from collections import namedtuple
    
    
    

    product =pd.read_csv (product, delimiter=";")
    Line = namedtuple('Line','Material Quantity unit_price row_total Del_Date')

    order_re = re.compile(r'PURCHASE ORDER NO. (.*)')
    material_re = re.compile(r'YOUR CODE: (.*)')
    line_re = re.compile("PC+")

    product['Material']=product['Material'].apply(str)

    lines = []

    with pdfplumber.open(file) as pdf:
            pages = pdf.pages
            for page in pdf.pages:
                text = page.extract_text()
                for line in text.split('\n'):
                    #print(line)
                    comp = order_re.search(line)
                    material = material_re.search (line)
                    if comp:
                        order_number = comp.group(1)

                    elif line_re.search(line):
                        items = line.split()

                    if material:
                        materialnumber =line.split()[2]
                        lines.append(Line(materialnumber, items[1],items[4],items[5],items[6]))


    df = pd.DataFrame(lines)
    df['Quantity']=df['Quantity'].astype('float')
    df['Quantity']=df['Quantity'].astype('int64')

    #complete= pd.merge(product, df,on='Material')

    df.to_excel(product+'.xlsx')
```
