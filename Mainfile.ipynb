```python
import re
import PyPDF2
from ipynb.fs.full.SLO import *
import pdfplumber
import os

product ='C:/Users/andreas.nederman/PDF/products.csv'

directory = "C:/Users/andreas.nederman/OneDrive - Glen Dimplex/ExceltoPDF/"
for file in os.listdir(directory):
    if file.endswith(".pdf"):
        file = directory+file
        pdf = pdfplumber.open(file)
        page = pdf.pages[0]
        text = page.extract_text()
        product= os.path.splitext(file)[0]
        print(product)
        
        if re.search('SLO SE', text):

            slo(file, product)
            print("SLO")

        elif re.search('Lemvigh-Müller', text):

            LM(file, product)
            print("LM")

        elif re.search('AHLSELL SVERIGE', text):

            AhlsellSE(file, product)
            print("AhlsellSE")

        elif re.search('Ahlsell Norge AS', text):

            AhlsellNO(file, product)
            print("Ahlsell Norge AS")        

        elif re.search('edmundson', text):

            edmundson(file, product)
            print("edmundson")  

        elif re.search('Rexel Sverige', text):

            RexelSE(file, product)
            print("RexelSE") 

        elif re.search('ELEKTROSKANDIA NORGE AS', text):

            ELEKTROSKANDIANO(file, product)
            print("ELEKTROSKANDIANO") 

        elif re.search('electric−center', text):

            edmundson(file, product)
            print("electric−center") 

        elif re.search('bewdirect', text):

            BEW(file, product)
            print("BEW")   

        elif re.search('SLO Oy', text):

            SLOFI(file, product)
            print("SLO Oy")   

        elif re.search('Stearn', text):

            StearnUK(file, product)
            print("StearnUK")  

        elif re.search('DHS', text):

            DHS(file, product)
            print("DHS")  
        
        elif re.search('Medlock Electrical', text):

            medloc(file, product)
            print("medloc")  

        else:

            print("NOT SET UP")
```
